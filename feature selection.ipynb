```python
from sklearn.feature_selection.variance_threshold import VarianceThreshold 
from sklearn.cluster import KMeans
from pandas import read_excel
```

```python
vt = VarianceThreshold()
```

```python
slo_df = read_excel('slo_dataset.xlsx')
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
```

```python
stats_df = slo_df.select_dtypes(include=numerics)
```

```python
stats_df.drop([col for col in stats_df.columns if 'Id' in col], axis=1, inplace=True)
```

```python
stats_df.drop(['gameDuration', 'season', 'champLevel'], axis=1, inplace=True)
```

```python
top_stats = stats_df[::5]
top_nan_cols = top_stats.iloc[:, top_stats.isnull().any().tolist()].columns
top_stats.drop(top_nan_cols, axis=1, inplace=True)
top_labels = slo_df[::5].win

jung_stats = stats_df[1::5]
jung_nan_cols = jung_stats.iloc[:, jung_stats.isnull().any().tolist()].columns
jung_stats.drop(jung_nan_cols, axis=1, inplace=True)
jung_labels = slo_df[1::5].win

mid_stats = stats_df[2::5]
mid_nan_cols = mid_stats.iloc[:, mid_stats.isnull().any().tolist()].columns
mid_stats.drop(mid_nan_cols, axis=1, inplace=True)
mid_labels = slo_df[2::5].win

adc_stats = stats_df[3::5]
adc_nan_cols = adc_stats.iloc[:, adc_stats.isnull().any().tolist()].columns
adc_stats.drop(adc_nan_cols, axis=1, inplace=True)
adc_labels = slo_df[3::5].win

supp_stats = stats_df[4::5]
supp_nan_cols = supp_stats.iloc[:, supp_stats.isnull().any().tolist()].columns
supp_stats.drop(supp_nan_cols, axis=1, inplace=True)
supp_labels = slo_df[4::5].win
```

```python
km = KMeans(n_clusters=3)
```

```python
clusters = km.fit_predict(X=top_stats)
```

```python
top_stats['clusters'] = clusters
```

```python
top_c_0 = top_stats[top_stats.clusters == 0]
top_c_1 = top_stats[top_stats.clusters == 1]
top_c_2 = top_stats[top_stats.clusters == 2]
```

```python
vt = VarianceThreshold()
```

```python
vt.fit(X=top_c_0)
```
